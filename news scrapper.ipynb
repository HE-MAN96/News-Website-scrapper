The vectorizer objects provided by Scikit-Learn allow us to perform conversion of categorical data to numerical form and even allowing preprocessing.Some of the vectorizer methods are given below:

1.) Count vectorizer: This is the most straightforward one,  it counts the number of times a token(word) shows up in the document and uses this value as its weight in the document term matrix.

2.) Hash Vectorizer: This one is designed to be as memory efficient as possible. Instead of storing the tokens as strings, the vectorizer applies the hashing trick(Using hash function) to encode them as numerical indexes. The downside of this method is that once vectorized, the features’ names can no longer be retrieved.

3.)TF-IDF Vectorizer:  TF-IDF stands for “term frequency-inverse document frequency”, meaning the weight assigned to each token not only depends on its frequency in a document but also how recurrent that term is in the entire document. 



## Q2:
Ans:3 things to evaluate a classificcation model:

1.) Confusion Matrix(F1 Score): It is a matrix representation of the results of any binary testing. It is composed of 4 columns namely:
~True positives(classifier predicted the positive case correctly)
~True negative(classifier predicted nagative case correctly)
~false positive(Classifier predicted positve case wrong)
~False negative(Classifier predicted negative case wrong)
based on the values of the above columns we can calculate Precision, recall and accuracy of the model.

2.)Jaccard Index
'JI' is defined as:
JI(Y ⋂ Y’)  = Y ⋂ Y’ / Y+Y’ - |Y⋂Y’|                Y = test/True set,   Y' = Predicted set
If 'JI' = 1.0, entire predicted set is exactly matching test set i.e., accuracy is 100%
if 'JI' = 0.0, accuracy is 0.
Closer the value of 'JI' to 1, the better is the accuracy.

3.)Logg Loss
Lograthmic loss measures the performance of a classifier where the predicted output/Class is a probablity value between 0 and 1.
values close to 0 indicates the probability that our model accurately predicts the test set is very low and will result in high logg loss.
Logg loss is calculated as:
  Logg Loss = 1/n {sigma of( Ylog(y' )+(1-y)log(1-y')}
  A low value of log loss indicates a good classifier model.

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
url = "https://inshorts.com/en/read"

r1 = requests.get(url)
coverpage = r1.content
soup1 = BeautifulSoup(coverpage)
#coverpage_news = soup1.find_all('h3', class_='fc-item__title')
soup1
body = soup1.find_all(attrs={"itemprop": "articleBody"},limit=25)
bodylen = len(body)
author = soup1.find_all(attrs={"class": "author"},limit=25)
authorlen= len(author)
time = soup1.find_all(attrs={"class": "time"},limit=25)
timelen = len(time)
date =soup1.find_all(attrs={"class": "date"},limit=25)
datelen = len(date)
#print(bodylen,authorlen,datelen,datelen)
soup1
```

```python
records = []
n=1
while n<25:
    content = body[n].get_text()
    authorName = author[n].get_text()
    Time = time[n].get_text()
    Date = date[n].get_text()
    records.append((content,authorName,Time,Date))
    n=n+1
    
#print(records)    


df=pd.DataFrame(records, columns=['Content', 'Author','Time', 'Date'])
#df.to_csv(r'C:\Users\heman\Desktop\data.csv')
df
```

### Now, we are trying to determine, given an article we predict its author out of the 7 distinct authors


```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

```python
print(df['Author'].unique())
x_train,x_test,y_train,y_test = train_test_split(df['Content'],df['Author'],test_size=0.2,random_state=1)
print("number of test samples :", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
text_clf = Pipeline([('vect', CountVectorizer()),
...                      ('tfidf', TfidfTransformer()),
...                      ('clf', MultinomialNB()),
... ])
text_clf.fit(x_train,y_train)
```

```python
predicted = text_clf.predict(x_test)
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, text_clf.predict(x_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, predicted))

```

```python
text_clf.fit(df['Content'],df['Author'])
pd1 = text_clf.predict(df['Content'])
print("final accuracy: ",metrics.accuracy_score(df['Author'],pd1))
```

### The dataset has only 25 rows, that's why splitting it into 80:20 ratio gives a test set accuracy of just 20%. Since our dataset is so small, test set accuracy is justified. But overall on the whole data set Naive bayes algorithm gives an accuracy of 79%.
we can try more algorithms too, but since dataset is too small, the accuracy would not be much. We can try to improve the accuracy by using 'cross validation' and 'Grid Search'. Trying Grid Search:

```python

from sklearn.model_selection import GridSearchCV
parameters = {'vect__ngram_range': [(1, 1), (1, 2)],'tfidf__use_idf': (True, False),'clf__alpha': (1e-2, 1e-3)}
```

```python
gs_clf = GridSearchCV(text_clf, parameters, n_jobs=-1)
gs_clf = gs_clf.fit(x_train, y_train)
```

```python
gs_clf.best_score_
gs_clf.best_params_
```

## using best parameters suggested by grid search

```python
text_clfnew = Pipeline([('vect', CountVectorizer()),
...                      ('tfidf', TfidfTransformer()),
...                      ('clf', MultinomialNB(alpha=0.01)),
... ])
text_clfnew.fit(x_train,y_train)
```

```python
predicted = text_clfnew.predict(x_test)
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, text_clfnew.predict(x_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, predicted))
```

The parameters suggested by grid search has significently improved the accuracy of the model.
